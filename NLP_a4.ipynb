```python
len(doc)
```

```python
from collections import defaultdict
from operator import itemgetter
```

```python
count_pos = defaultdict(int)
for token in doc:
    count_pos[token.pos_] += 1

count_pos
```

```python
count_tag = defaultdict(int)
for token in doc:
    count_tag[token.tag_] += 1

count_tag
```

```python
sorted(count_tag.items(), key=itemgetter(1), reverse=True)
```

```python
#Percentage
percentage_tag = {}
for k, v in count_tag.items():
    percentage_tag[k] = v/len(doc)
sorted(percentage_tag.items(), key=itemgetter(1), reverse=True)
```

## 50 most frequent noun lemmas

```python
#spaCy
text_clean = [w for w in doc if w.text not in jieba_stop_words]
noun_lemmas_spacy = defaultdict(int)
for i in text_clean:
    if i.tag_ in ['NN', 'NNP', 'NNB']:
        noun_lemmas_spacy[i.lemma_] += 1

sorted(noun_lemmas_spacy.items(), key = itemgetter(1), reverse=True)[:50]
```

```python
text_clean_jieba = [t for t in jieba_tokens if t[0] not in jieba_stop_words]
```

```python
#Jieba
noun_lemmas_jieba = defaultdict(int)
for i in text_clean_jieba:
    if i[1] in ['n', 'nr', 'nz', 'ns', 'ng', 'nt']:
        noun_lemmas_jieba[i[0]] += 1

sorted(noun_lemmas_jieba.items(), key = itemgetter(1), reverse=True)[:50]
```

## 50 most frequent verb lemmas

```python
#spaCy
verb_lemma_spacy = defaultdict(int)
for i in text_clean:
    if i.tag_ in ['VV', 'VC']:
        verb_lemma_spacy[i.lemma_] += 1
sorted(verb_lemma_spacy.items(), key = itemgetter(1), reverse=True)[:50]
```

```python
#jieba
v_lemmas_jieba = defaultdict(int)
for i in text_clean_jieba:
    if i[1] in ['vn', 'v', 'vg', 'vq']:
        v_lemmas_jieba[i[0]] += 1

sorted(v_lemmas_jieba.items(), key = itemgetter(1), reverse=True)[:50]
```

## 50 most frequent adj lemmas

```python
#spaCy
adj_lemma_spacy = defaultdict(int)
for i in text_clean:
    if i.tag_ in ['JJ']:
        adj_lemma_spacy[i.lemma_] += 1
sorted(adj_lemma_spacy.items(), key = itemgetter(1), reverse=True)[:50]
```

```python
#jieba
adj_lemmas_jieba = defaultdict(int)
for i in text_clean_jieba:
    if i[1] in ['a', 'ad', 'an', 'ag']:
        adj_lemmas_jieba[i[0]] += 1

sorted(adj_lemmas_jieba.items(), key = itemgetter(1), reverse=True)[:50]
```
